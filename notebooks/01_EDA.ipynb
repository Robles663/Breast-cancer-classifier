```python
# Cargamos librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/breast_cancer.csv")
```

```python
print(f"Shape: {df.shape}")
print(f"\nClases:\n{df['target'].value_counts()}")
print(f"\nValores nulos:\n{df.isnull().sum().sum()}")
print(f"\nEstadísticas descriptivas:")
df.describe()
```

```python
sns.countplot(x='target', data=df, hue='target', palette='Set2', legend=False)
plt.xticks([0, 1], ['Maligno', 'Benigno'])
plt.title('Distribución de clases')
plt.xlabel('Diagnóstico')
plt.ylabel('Cantidad de muestras')
plt.tight_layout()
plt.show()
```

```python
features = df.columns[:-1]  # todas menos target

fig, axes = plt.subplots(10, 3, figsize=(18, 40))
axes = axes.flatten()

for i, feature in enumerate(features):
    axes[i].hist(df[df['target'] == 0][feature], alpha=0.6, label='Maligno', color='tomato', bins=20)
    axes[i].hist(df[df['target'] == 1][feature], alpha=0.6, label='Benigno', color='steelblue', bins=20)
    axes[i].set_title(feature)
    axes[i].legend()

plt.tight_layout()
plt.show()
```

## Hipótesis preliminares

Basado en el análisis visual de las distribuciones:

1. Las caracteristicas del grupo `worst` muestran mejor separación entre clases que las del 
   grupo `mean`, siendo `worst concave points`, `worst perimeter` y `worst radius` 
   las más prometedoras para el modelo.

2. Caracteristicas como `symmetry`, `smoothness` y `fractal dimension` muestran poca 
   separación entre clases y podrían tener bajo poder predictivo individual.

3. Varias caracteristicas del grupo `error` presentan distribuciones fuertemente sesgadas 
   que podrían requerir transformación logarítmica en el preprocesamiento.

4. La separación visual sugiere que el problema es linealmente separable en varias 
   dimensiones, lo que favorece modelos como Logistic Regression como baseline.

```python
# Matriz de correlacion
corr = df.drop('target', axis= 1).corr()
plt.figure(figsize=(12, 10))  # Aumenta tamaño

sns.heatmap(
    corr,
    annot=False,
    cmap='coolwarm',
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": .8}
)

plt.xticks(rotation=45, ha='right')  # rota etiquetas
plt.yticks(rotation=0)

plt.tight_layout()  # ajusta todo automáticamente
plt.show()
```

## Análisis de correlación

- `mean radius`, `mean perimeter` y `mean area` forman un grupo de features 
  altamente redundantes (correlación > 0.9). Lo mismo ocurre con sus equivalentes 
  en el grupo `worst`. PCA en preprocesamiento ayudará a reducir esta redundancia.

- `mean concavity` y `mean concave points` están fuertemente correlacionadas, 
  ya que ambas miden aspectos similares de la irregularidad del contorno nuclear.

- `mean fractal dimension` muestra correlación baja o negativa con la mayoría 
  de features, lo que es consistente con su bajo poder de separación observado 
  en los histogramas.

- Las features del grupo `error` forman un bloque relativamente independiente, 
  con baja correlación respecto a `mean` y `worst`.

```python
top_features = [
    'mean radius',
    'mean concave points', 
    'worst radius',
    'worst perimeter',
    'worst concave points',
    'worst area',
    'target'
]

sns.pairplot(
    df[top_features],
    hue='target',
    palette={0: 'tomato', 1: 'steelblue'},
    plot_kws={'alpha': 0.5},
    diag_kind='kde'
)

plt.suptitle('Pairplot — Top predictive features', y=1.02)
plt.show()
```

## Análisis Pairplot

- La combinación `worst concave points` + `mean concave points` ofrece la mejor 
  separación visual entre clases, con benignos concentrados en valores bajos y 
  malignos en valores altos en ambas dimensiones.

- Las features del grupo `worst` (radius, perimeter, area) muestran relaciones 
  casi perfectamente lineales entre sí, confirmando la redundancia detectada en 
  el heatmap. PCA reducirá esta redundancia en el preprocesamiento.

- La mayoría de combinaciones muestran separación suficiente para sugerir que el 
  problema es linealmente separable en varias dimensiones, favoreciendo un buen 
  desempeño de Logistic Regression como baseline.

```python
# Separar features por grupo
mean_features = [col for col in df.columns if 'mean' in col]
se_features = [col for col in df.columns if 'error' in col]
worst_features = [col for col in df.columns if 'worst' in col]

fig, axes = plt.subplots(3, 1, figsize=(16, 18))

for ax, features, title in zip(
    axes,
    [mean_features, se_features, worst_features],
    ['Mean Features', 'Standard Error Features', 'Worst Features']
):
    df[features].boxplot(ax=ax)
    ax.set_title(title)
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Corregir el escalado
fig, axes = plt.subplots(3, 2, figsize=(18, 18))

grupos = [
    (mean_features, 'mean area', 'Mean Features — Sin área', 'Mean Features — Solo área'),
    (se_features, 'area error', 'SE Features — Sin área', 'SE Features — Solo área'),
    (worst_features, 'worst area', 'Worst Features — Sin área', 'Worst Features — Solo área'),
]

for i, (features, area_col, title_sin, title_con) in enumerate(grupos):
    sin_area = [f for f in features if f != area_col]
    
    # Sin la columna de área
    df[sin_area].boxplot(ax=axes[i, 0])
    axes[i, 0].set_title(title_sin)
    axes[i, 0].tick_params(axis='x', rotation=45)
    
    # Solo la columna de área
    df[[area_col]].boxplot(ax=axes[i, 1])
    axes[i, 1].set_title(title_con)

plt.tight_layout()
plt.show()
```

## Análisis de Outliers

- Las features de área (`mean area`, `area error`, `worst area`) presentan los 
  valores más extremos del dataset, con casos que duplican o triplican el rango 
  intercuartílico. Estos representan tumores de gran tamaño y son clínicamente 
  válidos — no se eliminan.

- `worst perimeter` y `perimeter error` también muestran outliers significativos, 
  consistente con su relación geométrica directa con el área.

- Decisión de preprocesamiento: se usará `RobustScaler` en lugar de `StandardScaler` 
  para reducir el impacto de los outliers en modelos sensibles a escala como 
  Logistic Regression y SVM.

- No se elimina ninguna muestra por outliers, ya que en contextos clínicos los 
  casos extremos suelen ser los más informativos para el modelo.
